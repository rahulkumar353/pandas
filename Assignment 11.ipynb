```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
```

```python
new=delivery.merge(data,left_on='match_id',right_on='id')
```

```python
# problem 1- To find the purple cap holder 
new.groupby(['season','bowler'])['player_dismissed'].count().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')
```

```python
# problem 2
# finding top batsmans who scored maximum runs
top_batsman=new.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
#finding top ten wicket takers
dismissal=['bowled','caught','caught and bowled','stumped','lbw','hit wicket']
wickets=new[new['dismissal_kind'].isin(dismissal)]
top_bowler=wickets.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(10)
```

```python
# top batsman dataframe
top_bat_data=(new[new['batsman'].isin(top_batsman.index.tolist())])
top_bat_data
```

```python
#top bowlers vs top batsman dataframe 
required_data=top_bat_data[top_bat_data['bowler'].isin(top_bowler.index.tolist())]
required_data
```

```python
pivot=required_data.pivot_table(index='bowler',columns='batsman',values='batsman_runs',aggfunc='sum')
pivot
```
