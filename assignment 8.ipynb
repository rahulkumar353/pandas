```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("matches.csv")
```

```python
data
```

```python
#problem 1
a=data['team1'].value_counts()
b=data['team2'].value_counts()
c=a+b
team=c.sort_values(ascending=False).head()
team.plot(kind='bar')
```

```python
#problem 2
def team_record(teama,teamb):
    
    mask1=data['team1']==teama
    mask2=data['team2']==teama
    mask3=data['team1']==teamb
    mask4=data['team2']==teamb
    return data[(mask1 | mask2) & (mask3 | mask4)]['winner'].value_counts()
```

```python
team_record('Mumbai Indians','Chennai Super Kings')
```
