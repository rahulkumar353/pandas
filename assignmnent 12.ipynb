```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
delivery=pd.read_csv('deliveries.csv')
data=pd.read_csv('matches.csv')
```

```python
new=delivery.merge(data,left_on='match_id',right_on='id')

```

```python
# problem 1
def most_runs(batsman):
    a=new[new['batsman']==batsman]
    b=a.groupby('season')['batsman_runs'].sum()
    b.plot(kind='line',marker='D')
```

```python
most_runs('MS Dhoni')
```

```python
# problem 2
def batsman_score(batsman):
    bat=new[(new['batsman']==batsman) & (new['batsman_runs']!=0)]
    a=bat.groupby('batsman_runs')['batsman_runs'].sum()
    a.plot(kind='pie',autopct='%0.1f%%',shadow=True)
    
```

```python
batsman_score('DA Warner')
```
